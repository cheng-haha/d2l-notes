```python
%matplotlib inline
import torch
from d2l import torch as d2l
```

# 探索一下relu，sigmoid,tanh

## 1. relu函数

```python
x = torch.arange( -8.0 , 8.0 , 0.1 ,requires_grad= True)
y = torch.relu( x )
d2l.plot( x.detach() , y.detach() , 'x' , 'relun(x)' , figsize=(5, 2.5))
```

```python
y.backward(torch.ones_like( x ) ,retain_graph=True  )#反向传播
d2l.plot( x.detach() , x.grad  , 'x' , 'the grad of relu' , figsize=(5, 2.5))
```

## 2.sigmoid函数

```python
x = torch.arange( -6.0 , 6.0 , 0.1 , requires_grad= True) 
y = torch.sigmoid( x )
d2l.plot( x.detach()  , y.detach() , ' x ',  ' y ' , figsize= (5,2.5)) 
```

```python
y.backward( torch.ones_like( x ) ,retain_graph= True  )
d2l.plot( x.detach() , x.grad , 'x' , 'the grad of x' , figsize=( 5,2.5 ) )
```

* sigmoid' = sigmoid( 1 - sigmoid )

## 3. tanh函数

```python
x.grad.data.zero_()
y  = torch.tanh( x ) 
y.backward( torch.ones_like( x ) ,retain_graph= True )
d2l.plot( x.detach()  ,x.grad , 'x' , ' the grad of x ' , figsize=( 5,2.5)) 
```
